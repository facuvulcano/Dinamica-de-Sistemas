Ecuaciones Diferenciales:

Las ecuaciones diferenciales son ecuaciones que relaacionan una funcion con sus derivadas. En terminos generales, se utilizan para describir fenomenos donde una cantidad varia en funcion de otra. Existen varios tipos de ecuaciones diferenciales, pero las mas comunes en la dinamica de sistemas son:

1- Ecuaciones diferenciales ordinarias (EDO): Involucran funciones de una variable independiente y sus derivadas. Se utilizan para modelar sistemas donde el cambio ocurre respecto al tiempo.

Las ecuaciones diferenciales lineales son un tipo especifico de ecuaciones diferenciales que son particularmente importantes debido a sus propiedades matematicas y su amplia aplicabilidad.

El orden de una ecuacion diferencial es el grado de la derivada mas alta presente en la ecuacion. En otras palabras, es el mayor numero de veces que la funcion desconocida ha sido derivada.

En el estudio de ecuaciones diferenciales y sistemas dinamicos, los puntos de equilibrio son estados donde el sistema no cambia, es decir, donde la tasa de cambio es cero. Estos puntos pueden clasificarse en diferentes tipos segun su comportamiento local, como sumideros, fuentes y puntos de silla.

Punto de equilbrio:

- Un punto de equilibrio (o punto fijo) de un sistema dinamico es un punto donde la derivada de la variable de estado es cero. Para un sistema descrito por xË™ = f(x), un punto xe es un punto de equilibrio si:

f(xe) = 0

Clasificacion de puntos de equilbrio:

- Sumidero:
    - Tambien conocido como punto de equilbrio estable
    - Cualquier perturbacion pequenia del estado llevaria al sistema de regreso al punto de equilibrio
    - Matricialmente, si los valores propios del sistema linealizado en el punto de equilibrio tienen partes reales negativas, el punto es un sumidero.
    - Visualmente, en un diagrama de fase, las trayectorias convergen hacia el sumidero.

- Fuente:
    - Tambien conocido como un punto de equilibrio inestable
    - Cualquier perturbacion pequenia del estado llevara al sistema a alejarse del punto de equilibrio
    - Matricialmente, si los valores propios del sistema linealizado en el punto de equilibrio tienen partes reales positivas, el punto es una fuente.
    - Visualmente, en un diagrama de fase, las trayectorias divergen desde la fuente.

- Punto de silla:
    - Un punto de equilibrio que es estable en una direccion y inestable en otra.
    - Matricialmente, si el sistema linealizado en el punto de equilbirio tiene valores propios con partes reales de diferentes signos, el punto es un punto de silla.
    - Visualmente, en un diagrama de fase, las trayectorias se aproximan al punto de silla en una direccion y se alejan en otra.
